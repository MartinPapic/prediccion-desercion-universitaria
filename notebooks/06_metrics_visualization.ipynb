# 6. Visualización de Métricas (Gráficos Comparativos)
Gráficos de barras para comparar de manera visual el Accuracy, Recall, Silhouette y Recompensas RL.

```python
import os
if os.getcwd().endswith('notebooks'):
    os.chdir('..')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
supervised_df = pd.read_csv("results/metrics/supervised_metrics.csv")
unsupervised_df = pd.read_csv("results/metrics/unsupervised_metrics.csv")
rl_df = pd.read_csv("results/metrics/rl_metrics.csv")
```

### Supervisados (Accuracy y Recall)

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Modelos Supervisados", fontsize=16)
sns.barplot(data=supervised_df, x="Model", y="Accuracy", ax=axes[0], hue="Model", legend=False, palette="viridis")
axes[0].tick_params(axis="x", rotation=45)
sns.barplot(data=supervised_df, x="Model", y="Recall", ax=axes[1], hue="Model", legend=False, palette="viridis")
axes[1].tick_params(axis="x", rotation=45)
plt.tight_layout()
plt.show()
```

### No Supervisados (Silhouette Score)

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=unsupervised_df, x="Model", y="Silhouette_Score", hue="Model", legend=False, palette="mako")
plt.title("Silhouette Score - No Supervisados")
plt.xticks(rotation=45)
plt.show()
```

### Aprendizaje por Refuerzo (RL) - Recompensa Total

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
rl_discrete = rl_df[rl_df["Model"].isin(["PPO", "A2C", "DQN"])]
rl_continuous = rl_df[rl_df["Model"].isin(["SAC", "DDPG"])]

sns.barplot(data=rl_discrete, x="Model", y="Total_Test_Reward", ax=axes[0], hue="Model", legend=False, palette="rocket")
axes[0].set_title("Recompensa Entorno Discreto (PPO, A2C, DQN)")

sns.barplot(data=rl_continuous, x="Model", y="Total_Test_Reward", ax=axes[1], hue="Model", legend=False, palette="magma")
axes[1].set_title("Error MSE Entorno Continuo (SAC, DDPG)")
plt.tight_layout()
plt.show()
```
